
#Netflix Data Analysis using pandas and matplotlib 


```python
import pandas as pd  
import matplotlib.pyplot as plt 
netflix_dataset = pd.read_csv(r"C:\\Users\\HP\\Downloads\\NetFlix.csv")
```

```python
df = netflix_dataset.copy()
df  
```

```python
#Data inspection 
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df["director"] = df["director"].fillna("N/A")
df["cast"] = df["cast"].fillna("N/A")
df["country"] = df["country"].fillna("N/A")
df.isna().sum()
```

```python
df.dropna(subset=["date_added","rating"],inplace=True)
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python

#What type of content Dominates Netflix: Movies or TV Shows?
df["type"].value_counts(normalize=True)*100
```

```python
df["type"].value_counts(normalize=True).plot(kind='pie', ylabel="", title="Content type Distributions")
plt.show()
```

```python
#Which contries produces the most Netflix content
df["country"].value_counts().sort_values(ascending=False).head(6).plot(kind='barh', ylabel="",title="Countries with Top Content on NETFLIX ")
plt.show()
```

```python
#What Genre are most common on Netflix 
df["genres"].value_counts().head().plot(kind='barh', xlabel="",ylabel="", title="Top Genres on Netflix"  )
plt.show()
```

```python
#How has Netflix's content library grown over time?

df["months"] = df["date_added"].dt.to_period("M").dt.to_timestamp()
df.head()
```

```python
df.groupby("months")["type"].size().plot(kind="line", xlabel=" ", ylabel=" ",title="Netflix Content Growth Over Time" , grid = True, figsize=(14,5) )
plt.show()
```

```python
content_by_time = df.groupby(["months","type"]).size().reset_index(name="count")
df.head()
```

```python
content_by_time = content_by_time.pivot(index="months", columns= "type", values="count")
```

```python
plt.Figure(figsize=(15,4))

plt.plot(content_by_time.index, content_by_time["Movie"],label=" Movie")
plt.plot(content_by_time.index, content_by_time["TV Show"],label=" Tv Shows")
plt.legend()
plt.title("Content Growth over time")
plt.grid(True)
plt.show()
```

```python
#What is the age distribution of Netflix Content (release year analysis)?

df["release_year"].hist(bins=10)
plt.show()
```

```python
df["release_year"].value_counts().plot(kind="bar",figsize=(15,4),xlabel="",title="Netflix Content Type Year of Production")
plt.show()
```
